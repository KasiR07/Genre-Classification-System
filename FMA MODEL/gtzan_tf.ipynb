```python
keras.backend.clear_session()
input_ = keras.layers.Input(shape=feat_shape[1:])
reshape = keras.layers.Reshape((feat_shape[1], feat_shape[2], -1))(input_)
hc1 = keras.layers.Conv2D(128, (4,4), activation="selu", kernel_initializer="lecun_normal", padding="same")(reshape)
po1 = keras.layers.MaxPooling2D(pool_size=(1,4))(hc1)
hc2 = keras.layers.Conv2D(128, (4,4), activation="selu", kernel_initializer="lecun_normal", padding="same")(po1)
hc3 = keras.layers.Conv2D(128, (4,4), activation="selu", kernel_initializer="lecun_normal", padding="same")(hc2)
hc4 = keras.layers.Conv2D(128, (4,4), activation="selu", kernel_initializer="lecun_normal", padding="same")(hc3)
po2 = keras.layers.MaxPooling2D(pool_size=(1,8))(hc3)
gp1 = keras.layers.Lambda(lambda x: [tf.reduce_mean(x, 1), tf.reduce_max(x, axis=1), tf.math.reduce_euclidean_norm(x, axis=1)])(po2)
ct1 = keras.layers.Concatenate()(gp1)
ft1 = keras.layers.Flatten()(ct1)
hd1 = keras.layers.Dense(1024, activation="selu", kernel_initializer="lecun_normal")(ft1)
dp1 = keras.layers.Dropout(0.3)(hd1)
hd2 = keras.layers.Dense(512, activation="selu", kernel_initializer="lecun_normal")(dp1)
dp2 = keras.layers.Dropout(0.3)(hd2)
hd3 = keras.layers.Dense(128, activation="selu", kernel_initializer="lecun_normal")(hd2)
dp3 = keras.layers.Dropout(0.3)(hd3)
output = keras.layers.Dense(10, activation="softmax")(dp3)
model = keras.Model(inputs=[input_], outputs=[output])
```

```python
def plot_spectrogram(spec, title=None, ylabel='freq_bin', aspect='auto', xmax=None):
    fig, axs = plt.subplots(1, 1)
    axs.set_title(title or 'Spectrogram (db)')
    axs.set_ylabel(ylabel)
    axs.set_xlabel('frame')
    im = axs.imshow(spec, origin='lower', aspect=aspect, cmap="winter")
    if xmax:
        axs.set_xlim((0, xmax))
    fig.colorbar(im, ax=axs)
    plt.show(block=False)
```
